```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from torch.nn.utils.parametrizations import weight_norm
import scipy as sp
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Load MNIST
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
train_dataset = datasets.MNIST(root="./data", train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root="./data", train=False, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1000)
```

```python
images, labels = next(iter(test_loader))
input_dim = images.shape[1]*images.shape[2]*images.shape[3]
```

```python
# Define w-val activation function
class MyActivation(nn.Module):
    def __init__(self, threshold: float):
        super().__init__()
        self.threshold = torch.as_tensor(threshold)
        self.entropy = None

    def forward(self,x):
        L = x.shape[1] if x.ndim == 2 else x.shape[1] * x.shape[2] * x.shape[3]
        alpha = torch.as_tensor(1/2, dtype=x.dtype, device=x.device)
        beta = torch.as_tensor((L-1)/2, dtype=x.dtype, device=x.device)

        # Normalize each weight vector going from a node in A to layer B
        x = F.normalize(x, p=2, dim=1)

        S_clamped = torch.clamp(x**2, min=1e-8, max=1-1e-8)

        # Compute the incomplete beta function manually using Beta distribution
        # This is done using a formula for the Beta CDF (this is a simple approximation)
        B = torch.exp(
            torch.lgamma(alpha + beta) - torch.lgamma(alpha) - torch.lgamma(beta)
        )  # Beta function normalization constant
        cdf = (S_clamped**(alpha - 1)) * ((1 - S_clamped)**(beta - 1)) / B
        beta_surv = 1 - cdf  # Survival function
        
        w_val = -torch.log10(beta_surv)
        assert w_val.shape == x.shape
        result = torch.where(w_val >= -torch.log10(self.threshold), x, 0)
        
        return result
```

```python
# 2. Define MLP with two hidden layers
class MLP(nn.Module):
    def __init__(self, threshold, input_dim=28*28, layer1=256, layer2=128, num_classes=10, activation_choice=nn.ReLU):
        super(MLP, self).__init__()

        # Handle activation selection
        def get_activation():
            return activation_choice() if activation_choice is not MyActivation else MyActivation(threshold)

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(input_dim, layer1),
            get_activation(),
            nn.Linear(layer1, layer2),
            get_activation(),
            nn.Linear(layer2, num_classes)
        )

    def forward(self, x):
        return self.classifier(x)

    def compute_layer_entropies(self):
        entropies = []
        for layer in self.classifier:
            if isinstance(layer, nn.Linear):
                entropy = compute_entropy_squared_weights(layer.weight)
                layer.entropy = entropy
                entropies.append(entropy)
        return entropies


def compute_entropy_squared_weights(weight_tensor):
    """Compute entropy using squared, normalized weights."""
    p = weight_tensor.pow(2).view(-1)
    p = p / (p.sum() + 1e-8)
    entropy = -torch.sum(p * torch.log(p + 1e-8)).item()
    return entropy
```

```python
# Setup
t = 6
activations = [nn.LeakyReLU(), nn.GELU(), nn.ReLU(), MyActivation(t)]
activation_names = ['LeakyReLU', 'GELU', 'ReLU', 'W-val activation']
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Logs
all_train_losses = []
test_accuracies = []
test_losses = []
entropy_per_activation = []  # <== NEW
models = []

# Training and Evaluation Loop
for act, name in zip(activations, activation_names):
    print(f"\nTraining with {name}...")
    model = MLP(threshold=t, input_dim=input_dim, activation_choice=type(act)).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.01)

    # Training
    per_epoch_losses = []
    for epoch in range(10):
        model.train()
        total_loss = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        avg_loss = total_loss / len(train_loader)
        per_epoch_losses.append(avg_loss)
        print(f"Epoch {epoch+1}, Loss: {avg_loss:.4f}")

    all_train_losses.append(per_epoch_losses)
    models.append(model)

    # Evaluation
    model.eval()
    correct = total = total_test_loss = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_test_loss += loss.item()
            preds = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    accuracy = correct / total
    avg_test_loss = total_test_loss / len(test_loader)
    test_accuracies.append(accuracy)
    test_losses.append(avg_test_loss)
    print(f"Test Accuracy: {accuracy:.4f}, Test Loss: {avg_test_loss:.4f}")

    # Entropy
    entropies = model.compute_layer_entropies()
    entropy_per_activation.append(entropies)
```

```python
# Plot training losses
plt.figure(figsize=(10, 5))
for losses, name in zip(all_train_losses, activation_names):
    plt.plot(losses, label=name)
plt.title("Training Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Plot test accuracies
plt.figure(figsize=(6, 4))
plt.bar(activation_names, test_accuracies, color='skyblue')
plt.ylabel("Test Accuracy")
plt.title("Final Test Accuracy by Activation")
plt.ylim(0, 1.0)
plt.grid(axis='y')
plt.show()

# Plot test losses
plt.figure(figsize=(6, 4))
plt.bar(activation_names, test_losses, color='salmon')
plt.ylabel("Test Loss")
plt.title("Final Test Loss by Activation")
plt.grid(axis='y')
plt.show()

# Plot entropy per layer
plt.figure(figsize=(10, 5))
for entropies, name in zip(entropy_per_activation, activation_names):
    plt.plot(range(1, len(entropies) + 1), entropies, marker='o', label=name)
plt.title("Entropy per Layer (Squared Weights)")
plt.xlabel("Layer Index")
plt.ylabel("Entropy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def plot_weight_distribution(model, threshold=1e-5):
    all_weights = []
    near_zero_count = 0
    total_count = 0

    # Iterate over model parameters
    for name, param in model.named_parameters():
        if 'weight' in name and param.requires_grad:
            weights = param.data.cpu().numpy().flatten()
            all_weights.extend(weights)
            # Count near-zero weights
            near_zero_count += (abs(weights) < threshold).sum()
            total_count += weights.size

            # Plot histogram for this layer
            plt.figure()
            plt.hist(weights, bins=100, alpha=0.7)
            plt.title(f'Weight distribution: {name}')
            plt.xlabel('Weight value')
            plt.ylabel('Frequency')
            plt.grid(True)
            plt.show()

    # Summary across all layers
    all_weights = torch.tensor(all_weights)
    overall_sparsity = (abs(all_weights) < threshold).sum().item() / all_weights.numel() * 100

    print(f"Overall near-zero weights (< {threshold}): {overall_sparsity:.2f}% of total weights")
```
